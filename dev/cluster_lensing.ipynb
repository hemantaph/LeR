```python
import numpy as np
import lensprop as lp
import cosmology as cc
from constants import *

# Initialising cosmology for mass function, distances, etc. 
aum_cosm = cc.cosmology(Om0,Omk,w0,wa,Ob,h_H0,Tcmb0,sigma8,ns,np.log10(8.0),1.0)
```

```python
Msun = 1.989e30
params = dict(
    zlmin = 0.2,
    zlmax = 1.7,
    Mmin = 1e10, # I am not sure this should be in log10 or not
    Mmax = 1e15,
    aum_cosm = aum_cosm,
)
```

```python
num_ = []
M200_arr = np.linspace(1e10,1e16,100)
zl = 1

for m in M200_arr:
    num_.append(class_.get_halomassfunc(m, zl))
```
