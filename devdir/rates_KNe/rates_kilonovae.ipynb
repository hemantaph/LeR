```python
import numpy as np
import matplotlib.pyplot as plt

# Constants given in the problem
M_lambda_0 = -16.3
Delta_M_lambda = 4
theta_0 = np.deg2rad(60)  # 60 degrees in radians

# Range of theta_v from 0 to 90 degrees
theta_v = np.linspace(0, np.pi/2, 400)  # radians

# Range for delta_M_lambda
delta_M_lambda = np.linspace(-1, 1, 3)  # taking three values for illustration: -1, 0, 1

# Function M_lambda_theta_v
def M_lambda_theta_v(theta_v, M_lambda_0, Delta_M_lambda, delta_M):
    return M_lambda_0 + Delta_M_lambda * ((1 - np.cos(theta_v)) / (1 - np.cos(theta_0))) + delta_M

# Calculate magnitude for different delta_M_lambda
magnitudes = [M_lambda_theta_v(theta_v, M_lambda_0, Delta_M_lambda, dm) for dm in delta_M_lambda]

# Plotting
plt.figure(figsize=(4, 4))
for mag, dm in zip(magnitudes, delta_M_lambda):
    plt.plot(np.degrees(theta_v), mag, label=f'$\delta M_{{\lambda}}$ = {dm}')

plt.title('Peak Absolute Magnitude $M_{\lambda, \\theta_v}$ vs. Viewing Angle $\\theta_v$')
plt.xlabel('Viewing Angle $\\theta_v$ (degrees)')
plt.ylabel('Peak Absolute Magnitude $M_{\lambda, \\theta_v}$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Constants given in the problem
M_lambda_0 = -16.3
Delta_M_lambda = 7
theta_0 = np.deg2rad(60)  # 60 degrees in radians

# Range of theta_v from 0 to 90 degrees
theta_v = np.linspace(0, np.pi/2, 400)  # radians

# Range for delta_M_lambda
delta_M_lambda = np.linspace(-1, 1, 3)  # taking three values for illustration: -1, 0, 1

# Function M_lambda_theta_v
def M_lambda_theta_v(theta_v, M_lambda_0, Delta_M_lambda, delta_M):

    result = np.zeros_like(theta_v)
    for i, theta in enumerate(theta_v):
        if theta<=theta_0:
            result[i] = M_lambda_0 + Delta_M_lambda * ((1 - np.cos(theta)) / (1 - np.cos(theta_0))) + delta_M
        else:
            result[i] = M_lambda_0 + Delta_M_lambda + delta_M
    return result

# Calculate magnitude for different delta_M_lambda
magnitudes = [M_lambda_theta_v(theta_v, M_lambda_0, Delta_M_lambda, dm) for dm in delta_M_lambda]

# Plotting
plt.figure(figsize=(10, 6))
for mag, dm in zip(magnitudes, delta_M_lambda):
    plt.plot(np.degrees(theta_v), mag, label=f'$\delta M_{{\lambda}}$ = {dm}')

plt.title('Peak Absolute Magnitude $M_{\lambda, \\theta_v}$ vs. Viewing Angle $\\theta_v$')
plt.xlabel('Viewing Angle $\\theta_v$ (degrees)')
plt.ylabel('Peak Absolute Magnitude $M_{\lambda, \\theta_v}$')
plt.legend()
plt.grid(True)
plt.show()
```
